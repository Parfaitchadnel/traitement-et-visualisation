```python
#1 : Générer des données synthétiques avec make_blobs
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Création des données
X, labels_true = make_blobs(n_samples=500, centers=3, cluster_std=0.5, random_state=42)

plt.scatter(X[:,0], X[:,1], c='blue', edgecolor='k', s=40)
plt.title("Données initiales")
plt.show()
```

```python
#2 : Appliquer K-Means et DBSCAN
from sklearn.cluster import KMeans, DBSCAN

# K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)
```

```python
#3 : Visualiser les clusters obtenus
fig, axes = plt.subplots(1, 2, figsize=(14,6))

# Visualisation K-Means
axes[0].scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', edgecolor='k', s=40)
axes[0].set_title("Clusters K-Means")

# Visualisation DBSCAN
axes[1].scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis', edgecolor='k', s=40)
axes[1].set_title("Clusters DBSCAN")

plt.show()
```

```python
#4
K-Means force chaque point à appartenir à un cluster, formant ainsi des groupes compacts.
DBSCAN identifie des clusters en fonction de la densité des points. Les points trop éloignés 
sont classifiés comme bruit (-1).
```
