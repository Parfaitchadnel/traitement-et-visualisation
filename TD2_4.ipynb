```python
#1 Charger le dataset Mall_Customers.csv
import pandas as pd

df_mall = pd.read_csv('Mall_Customers.csv')
df_mall.head()
```

```python
#2 Appliquer clustering hiérarchique et GMM
from sklearn.cluster import AgglomerativeClustering
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

# Utilisation des colonnes "Annual Income" et "Spending Score"
X = df_mall[['Annual Income (k$)', 'Spending Score (1-100)']]

# Clustering hiérarchique
hierarchical = AgglomerativeClustering(n_clusters=5)
df_mall['Hierarchical_Cluster'] = hierarchical.fit_predict(X)

# Gaussian Mixture Model (GMM)
gmm = GaussianMixture(n_components=5, random_state=42)
df_mall['GMM_Cluster'] = gmm.fit_predict(X)
```

```python
#3 Comparer les résultats (visualisation)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].scatter(X.iloc[:, 0], X.iloc[:, 1], c=df_mall['Hierarchical_Cluster'], cmap='rainbow')
axes[0].set_title('Clustering Hiérarchique')
axes[0].set_xlabel('Annual Income (k$)')
axes[0].set_ylabel('Spending Score (1-100)')

axes[1].scatter(X.iloc[:, 0], X.iloc[:, 1], c=df_mall['GMM_Cluster'], cmap='rainbow')
axes[1].set_title('Gaussian Mixture Model')
axes[1].set_xlabel('Annual Income (k$)')
axes[1].set_ylabel('Spending Score (1-100)')

plt.show()
```

```python
#4
GMM modélise les clusters comme des distributions gaussiennes, permettant :

Des clusters de formes variées (elliptiques par exemple).

Une estimation probabiliste (niveau de confiance pour chaque cluster).
```
